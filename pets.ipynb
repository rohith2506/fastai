```python
data = ImageDataBunch.from_name_re(images_path, fnames, pat, ds_tfms=get_transforms(), size=224, bs=bs).normalize(imagenet_stats)
```

```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

```python
data.show_batch(rows=1, figsize=(7, 6))
```

```python
data.classes
```

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.model
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage-1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)
losses, idxes = interp.top_losses()
```
